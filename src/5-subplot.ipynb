### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
import matplotlib.pyplot as plt

# Data for all plots
x = [1, 2, 3, 4, 5]
y_line = [i**2 for i in x]
y_bar = [5, 7, 3, 9, 6]
y_scatter = [2, 4, 6, 8, 10]
hist_data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5]
pie_labels = ['Python', 'Java', 'C++', 'JavaScript']
pie_sizes = [40, 25, 20, 15]

# Create 5 subplots in a 3x2 grid (with one unused)
fig, axs = plt.subplots(3, 2, figsize=(12, 10))

# Flatten axs to index easily
axs = axs.flatten()

# Line chart
axs[0].plot(x, y_line, marker='o', color='blue')
axs[0].set_title('Line Chart')

# Bar chart
axs[1].bar(x, y_bar, color='orange')
axs[1].set_title('Bar Chart')

# Scatter plot
axs[2].scatter(x, y_scatter, color='green')
axs[2].set_title('Scatter Plot')

# Histogram
axs[3].hist(hist_data, bins=5, color='skyblue', edgecolor='black')
axs[3].set_title('Histogram')
axs[3].set_xlabel('Value')
axs[3].set_ylabel('Frequency')

# Pie chart
axs[4].pie(pie_sizes, labels=pie_labels, autopct='%1.1f%%', startangle=140)
axs[4].set_title('Pie Chart')
axs[4].axis('equal')  # Equal aspect ratio for pie chart

# Hide the sixth (empty) subplot
fig.delaxes(axs[5])

# Adjust layout
plt.tight_layout()
plt.show()
```
